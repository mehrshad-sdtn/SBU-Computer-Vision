```python
import cv2
import matplotlib.pyplot as plt

def imshow(image):
   plt.imshow(image, cmap="gray")
```

```python
image = cv2.imread('saadatinia.jpg')  
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
imshow(gray_image)
```

```python
blur = cv2.GaussianBlur(gray_image, (37, 37), 0)
imshow(blur)
```

```python
sketch = cv2.divide(gray_image, blur, scale=255.0)
out = sketch  #cv2.GaussianBlur(sketch, (3, 3), 0)
imshow(out)
cv2.imwrite("sketch_mehrshad.jpg", out)
```
